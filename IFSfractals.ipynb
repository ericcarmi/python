```python
# Import the things we need
from pylab import plt,array,linspace,int8,zeros
import ITS
from IFSstructures import *
```

```python
A,T,p = getIFS("fern") # Get IFS parameters 
npdf = linspace(0,len(p)-1,len(p),dtype=int8)
M = 100 # Number of points
w = zeros((M))*1j
z0 = array(( [0],[0]))
for k in range(M):
    r = ITS.getSample(p,npdf)
    z0 = A[r] @ z0 + T[r]
    w[k] = z0[0] + 1j*z0[1]
plt.scatter(w.real,w.imag,s=1.0)
plt.show()
```
